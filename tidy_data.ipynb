Load the attendance.csv file and calculate an attendnace percentage for each student. One half day is worth 50% of a full day, and 10 tardies is equal to one absence.

```python
attendance = pd.read_csv("untidy-data/attendance.csv")
attendance.head()
```

```python
attendance = attendance.melt(id_vars = ["Unnamed: 0"], var_name = "date", value_name = "attendance_code").rename(columns = {"Unnamed: 0" : "student_name"})
attendance
```

```python
attendance.info()
```

```python
attendance.date.str.split("-", expand = True)
```

```python
attendance[["year", "month", "day"]] = attendance.date.str.split("-", expand = True)
attendance
```

```python
attendance = attendance[["student_name", "attendance_code"]]
attendance
```

```python
attendance.insert(2, "code_count", True)
```

```python
attendance
```

```python
attendance = attendance.groupby(by = ["student_name", "attendance_code"]).count().reset_index()
attendance
```

```python
attendance = attendance.pivot_table(index = "student_name", columns = "attendance_code", values = "code_count").reset_index()
attendance
```

```python
attendance = attendance.fillna(0)
attendance
```

```python
attendance = attendance[["student_name","P", "H", "T", "A"]]
attendance[["P", "H", "T", "A"]]
attendance
```

```python
attendance.info()
```

```python
attendance["total_sessions"] = attendance.sum(axis = 1)
attendance
```

```python
# def add_one(column):
#     for value in values:
#         if value >= 10:
#             value = 1
#         else: 


attendance['total_absences'] = attendance[["A", "T"]].sum(axis = 1).where(attendance['T'] >= 10, attendance["A"])
attendance
```

```python
attendance.info()
```

```python
attendance["H"] = attendance["H"]*(.5)
attendance['total_attendance'] = attendance[["P", "H", "T"]].sum(axis = 1)#.where(attendance['T'] < 10)
attendance
```

```python
attendance = attendance[["student_name", "total_attendance", "total_absences", "total_sessions"]]
attendance
```

```python
attendance["attendance_rate"] = attendance["total_attendance"] / attendance["total_sessions"]
attendance
```

#### 3. Cake Recipes

a. Read the cake_recipes.csv data. This data set contains cake tastiness scores for combinations of different recipes, oven rack positions, and oven temperatures.

b. Tidy the data as necessary.

c. Which recipe, on average, is the best? recipe b

d.Which oven temperature, on average, produces the best results? 275

e. Which combination of recipe, rack position, and temperature gives the best result? recipe b, bottom rack, 300 degrees


```python
cake_recipes = pd.read_csv("untidy-data/cake_recipes.csv")
cake_recipes
```

```python
cake_recipes = cake_recipes.melt(id_vars = "recipe:position", var_name = "oven_temp", value_name = "tastiness_score")
cake_recipes
```

```python
cake_recipes.info()
```

```python
cake_recipes[["recipe", "oven_position"]] = cake_recipes["recipe:position"].str.split(":", expand = True)
cake_recipes
```

```python
cake_recipes.columns
```

```python
cake_recipes = cake_recipes[['recipe', 'oven_position', 'oven_temp', 'tastiness_score']]
cake_recipes
```

c. Which recipe, on average, is the best? recipe b

```python
cake_recipes.groupby(by = "recipe").mean().sort_values(by = "tastiness_score", ascending = False).head(1)
```

d. Which oven temperature, on average, produces the best results? 275

```python
cake_recipes.groupby(by = "oven_temp").mean().sort_values(by = "tastiness_score", ascending = False).head(1)
```

e. Which combination of recipe, rack position, and temperature gives the best result? recipe b, bottom rack, 300 degrees

```python
cake_recipes.groupby(by = ["recipe", "oven_position", "oven_temp"]).mean().sort_values(by = "tastiness_score", ascending = False).head(1)
```
